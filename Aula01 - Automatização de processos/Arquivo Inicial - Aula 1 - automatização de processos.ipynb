```python
# 4 - Calcular o indicativo 
import pandas as pd #Muito utilizada para base de dados

tabela = pd.read_excel(r"C:\Users\Matheus\Intensivo-python\aula01\Vendas - Dez.xlsx") #r" -> rowstring *não interpreta caracteres especiais
display(tabela)

faturamento = tabela["Valor Final"].sum()
quantidades_prod_vendidos = tabela["Quantidade"].sum()
#display(faturamento)
#print(quantidades_prod_vendidos)
```

```python
# 5 - enviar email para a diretoria com os indicadores
pyautogui.hotkey("ctrl", "t")
time.sleep(2)
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")
time.sleep(12)

#clica em escrever
pyautogui.click(x=82, y=174)

#escrever destinatário
destinatario = "moisesjlima@outlook.com"
time.sleep(12)
pyautogui.write(destinatario)
pyautogui.press("tab")
#assunto
pyautogui.press("tab")
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")

#escrever o corpo do email
pyautogui.press("tab")
texto_corpo = f"""
Prezados, bom dia

O faturamento de ontem foi R${faturamento:,.2f}
A quantidades de produtos vendidas foi de {quantidades_prod_vendidos:,}

Abs,
Moises lima 
"""

pyperclip.copy(texto_corpo)
pyautogui.hotkey("ctrl", "v")

#enviar email
time.sleep(2)
pyautogui.hotkey("ctrl", "enter")

#fechar aba do gmail
time.sleep(8)
pyautogui.hotkey("ctrl", "w")
```
